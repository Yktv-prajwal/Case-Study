```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_style('darkgrid') # Clean Background
sns.set_palette("viridis") #Color
plt.rcParams["figure.figsize"] = (20,10) # Size
# ['white', 'dark', 'whitegrid', 'darkgrid', 'ticks']
```

```python
# Basic Distribution - Price Sprade Out

sns.histplot(data=df,x = "Price_Lakhs",kde = True, color = "Green", bins=30) # kde - Kernel Density Estimation (Slope Line)
plt.title("Distribution - Price Sprade Out")
plt.xlabel("Price")
plt.ylabel("Number of House")
plt.show()
```

```python
#BHK Most Expensive ? - Compare Price

sns.boxplot(data=df,x ="BHK", y = "Price_Lakhs", palette = "Set1", hue = "BHK")
plt.title("HOUSE price- Number of BHK")
plt.xlabel("BHK")
plt.ylabel("Price")
plt.show()
```

```python
# Scatter Plot - Relationship - Price vs Area

sns.scatterplot(data = df, x = "Area_sqft", y = "Price_Lakhs", hue="Locality", style="BHK", size="BHK",palette = "tab10")
plt.title("HOUSE Price Vs Area")
plt.xlabel("Area sqft")
plt.ylabel("Price")
plt.legend(bbox_to_anchor = (1.05,1), loc = "upper left")
plt.show()
```

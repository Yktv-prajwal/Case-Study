```python
# Load the dataset into a Pandas DataFrame and display the first 5 rows.
import pandas as pd

df = pd.read_csv("cc.csv")
df.head()
```

```python
# Check the shape, column names, and summary statistics of the dataset.
print(df.shape) #(row,col)
print(df.columns.to_list())
print('summary statistics \n ',df.describe()) # include='all'
```

```python
#check dtype
df.dtypes
```

```python
# Identify and handle missing values (fill or drop based on the data type).
print(df.isnull().sum())
# Convert Transaction_Date into datetime format and extract year, month, and day as new columns.
```

```python
# fill the missing value 
df['Merchant'] = df['Merchant'].fillna('Unknown')
df['Transaction_Type'] = df['Transaction_Type'].fillna('Unknown')
df['Category'] = df['Category'].fillna('Unknown')
df['Payment_Mode'] = df['Payment_Mode'].fillna('Unknown')
df['Transaction_Status'] = df['Transaction_Status'].fillna('Unknown')

# fill int values
df['Amount'] = df['Amount'].fillna(df['Amount'].median()) # mean()
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
# Convert Transaction_Date into datetime format and extract year, month, and day as new columns.

df.dtypes
```

```python
df['Transaction_Date']  = pd.to_datetime(df['Transaction_Date'],errors='coerce')
print(df.dtypes)
```

```python
#  year, month, and day as new columns
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day

print(df.info())
```

```python
df.head()
df.to_csv("clean_data.csv")
```

```python
# Data Selection and Indexing
# Retrieve all transactions made in January 2024.
# Find transactions where Amount > 1000 and Transaction_Type is "Online".
# Select only Approved transactions from the dataset.
df.head()
```

```python
# # Retrieve all transactions made in January 2023.
jan_24 = df[(df['Year'] == 2023) & ( df['Month'] == 1)] 
# print(jan_24.count())
jan_24.to_csv("jan_24.csv",index=False)
```

```python
# Q1 - Find transactions where Amount > 3000 and Transaction_Type is "Online".
high_transactions = df[(df['Amount'] > 3000) & (df['Transaction_Type'] == "Online")]
high_transactions.head()
```

```python
# Q2 - Select only Approved transactions from the dataset.
approved_transactions = df[df['Transaction_Status'] == 'Approved']
approved_transactions.head()
```

```python
# Data Manipulation and Feature Engineering
# Create a new column Discounted_Amount, assuming a 5% discount on all transactions above $4500.
df['Discounted_Amount'] = df['Amount'].apply(lambda x : x * 0.95 if x > 4500 else x)
df[['Amount','Discounted_Amount']].head()
```

```python
df.head()
```

```python
# Categorize the Transaction_Amount into Low, Medium, and High based on:
# Low: Below $2000
# Medium: Between $2000 - $4500
# High: Above $4500

df['Amount_Category'] = pd.cut(df['Amount'],
                               bins=[-float('inf'),2000,4500,float('inf')],
                               labels=['Low','Medium','High'])

df[['Amount','Amount_Category']].head()

# Drop the Merchant column if more than 30% of values are missing.
missing_value = df['Merchant'].isnull().mean()
print(missing_value)

if missing_value > 0.3:
    df.drop(columns=['Merchant'],inplace=True)
    print("Delete done ...")
else:
    print("no missing value found...")
```

```python
# Aggregation and Insights
# Find the total transaction amount per Category.
total_amount_per_category = df.groupby('Category')['Amount'].sum().reset_index()
total_amount_per_category
```

```python
# Determine the number of declined transactions per Payment_Mode.
decliend_transactions = df[df['Transaction_Status']=='Declined'].groupby('Payment_Mode').size().reset_index(name='declined_count')
decliend_transactions
```

```python
# Find the average transaction amount per Location.
avg_amount_loc = df.groupby('Location')['Amount'].mean().reset_index()
avg_amount_loc
```

```python
customer_info= pd.read_csv("customer_details.csv")
customer_info.head()
```

```python
merge_df = pd.merge(df,customer_info,on = "Customer_ID", how = "inner")

merge_df.head()
```

```python
Age_bin = [0,18,30,50,100]
age_labels = ["<18","18-30","30-50",">50"]

merge_df['Age_Group'] = pd.cut(merge_df['Age'],bins=Age_bin,labels=age_labels,include_lowest = True)
avg_amount_per_age_group = merge_df.groupby("Age_Group")["Amount"].mean().reset_index()
avg_amount_per_age_group
```

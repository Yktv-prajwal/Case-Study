```python
import pandas as pd
```

```python
'''  1. Data Exploration and Cleaning
Load the dataset into a Pandas DataFrame and display the first 5 rows.
Check the dataset shape, column names, data types, and summary statistics.
Identify missing values in each column.
Handle missing values appropriately:
Numerical columns: fill with mean or median.
Categorical columns: fill with mode or drop.
Convert Order_Date into datetime format.
Extract Year, Month, and Day from Order_Date into new columns. '''
```

```python
# Load the dataset into a Pandas DataFrame and display the first 5 rows.

data=pd.read_csv("rs.csv")
data.head()
```

```python
# Check the dataset shape, column names, data types, and summary statistics.

data.shape
data.info()
data.describe()
```

```python
# Identify missing values in each column.

print(data.isnull().sum())
```

```python
# Numerical columns: fill with mean or median.

data["Unit_Price"] = data["Unit_Price"].fillna(data["Unit_Price"].median())
data["Product"] = data["Product"].fillna(data["Product"])
data["Category"] = data["Category"].fillna(data["Category"])
data["Payment_Method"] = data["Payment_Method"].fillna(data["Payment_Method"])
data["Order_Status"] = data["Order_Status"].fillna(data["Order_Status"])
data["City"] = data["City"].fillna(data["City"])

print(data.isnull().sum())
```

```python

print(data.isnull().sum())
```

```python
# Convert Order_Date into datetime format.

data['Order_Date']  = pd.to_datetime(data['Order_Date'],errors='coerce')
print(data.dtypes)
```

```python
#  year, month, and day as new columns
data['Year'] = data['Order_Date'].dt.year
data['Month'] = data['Order_Date'].dt.month
data['Day'] = data['Order_Date'].dt.day

print(data.info())
```

```python
data.head()
```

```python
# 2. Data Selection and Filtering
# Retrieve all orders placed in March 2024.
# Find all orders where:
# Total_Amount > 1000
# Category is Electronics.
# Select only Completed orders.
# Retrieve orders placed from Mumbai or Delhi.
```

```python
# Retrieve all orders placed in March 2024.

mar_24 = data[(data["Year"] == 2024) & (data["Month"] == 3)]

print(mar_24.count())
```

```python
# Total_Amount > 1000

Total_Amount  = data[(data["Total_Amount"] > 1000 )]

Total_Amount
```

```python
# Category is Electronics.

Category  = data[(data["Category"] == "Electronics" )]

Category
```

```python
# Select only Completed orders.

Orders  = data[(data["Order_Status"] == "Completed" )]

Orders
```

```python
# Retrieve orders placed from Mumbai or Delhi.

mh_dl = data[(data["City"] == "Mumbai") | (data["City"] == "Delhi")]

mh_dl
```

```python
# 3. Data Manipulation and Feature Engineering
# Create a new column Discounted_Amount:
# Apply a 10 percent discount on orders where Total_Amount > 800.
# Create a column Price_Level based on Total_Amount:
# Low: Below 200
# Medium: 200 to 800
# High: Above 800
# Create a new column Revenue_Per_Item:
# Revenue_Per_Item = Total_Amount / Quantity
# Drop the Product column if more than 40 percent of values are missing.
```

```python
# Create a new column Discounted_Amount:
# Apply a 10 percent discount on orders where Total_Amount > 800.

data['Discounted_Amount'] = data['Total_Amount'].apply(lambda x : x * 0.90 if x > 800 else x)
data[['Total_Amount','Discounted_Amount']].head()
```

```python
data.head()
```

```python
# Create a column Price_Level based on Total_Amount:
# Low: Below 2000
# Medium: 2000 to 4500
# High: Above 4500

data['Amount_Category'] = pd.cut(data['Total_Amount'],
                               bins=[-float('inf'),2000,4500,float('inf')],
                               labels=['Low','Medium','High'])

data[['Total_Amount','Amount_Category']].head()
```

```python
# Create a new column Revenue_Per_Item:
# Revenue_Per_Item = Total_Amount / Quantity

data["Revenue_Per_Item"] = data["Total_Amount"] / data["Quantity"]
data[["Revenue_Per_Item", "Total_Amount"]].head()
```

```python
# Drop the Product column if more than 40 percent of values are missing.

if data["Product"].isnull().mean() > 0.40:
    data.drop(columns=["Product"], inplace=True)
```

```python
# 4. Aggregation and Business Insights
# Find the total revenue per Category.
# Calculate the average order value per Payment_Method.
# Identify the top 5 cities by total sales.
# Find the number of cancelled and returned orders per Category.
# Determine the monthly sales trend.
```

```python
# Find the number of cancelled and returned orders per Category.

cancle_Return = data[data["Order_Status"].isin(["Cancelled", "Returned"])]

ReturnCal = cancle_Return.groupby(["Category", "Order_Status"]).size()

ReturnCal.reset_index()
```

```python
# 5. Customer Behavior Analysis
# Identify customers who placed more than 5 orders in a single month.
# Find customers whose average order value exceeds 1500.
# Identify customers who placed orders from multiple cities.
# Find repeat customers with more than one order.
```

```python
# Identify customers who placed more than 5 orders in a single month.

Cust_ord = data.groupby(["Customer_ID", "Month"])["Order_ID"].count().reset_index(name="Orders")

Cust_ord = Cust_ord[Cust_ord["Orders"] > 5]

Cust_ord
```

```python
data.head()
```

```python
# Find customers whose average order value exceeds 1500.

avg_value = data.groupby("Customer_ID")["Total_Amount"].mean().reset_index(name="Avg_Order_Value")

avg_value = avg_value[avg_value["Avg_Order_Value"] > 1500]

avg_value.head(11)
```

```python
# Identify customers who placed orders from multiple cities.

Customers_From_Mul_cities = data.groupby("Customer_ID")["City"].unique().reset_index()
Customers_From_Mul_cities = Customers_From_Mul_cities[Customers_From_Mul_cities["City"].apply(len) > 1]
Customers_From_Mul_cities.head(11)
```

```python
# Find repeat customers with more than one order.

Rep_cust = data.groupby("Customer_ID")["Order_ID"].size().reset_index(name = "Orders")

Rep_cust = Rep_cust[Rep_cust["Orders"] > 1]
Rep_cust.head(11)
```

```python
# 7. Data Merging and Joining
# You are given another dataset named customer_details.csv with the following columns:

# Column Name	Description
# Customer_ID	Unique customer identifier
# Age	Customer age
# Gender	Male or Female
# Membership_Level	Silver, Gold, or Platinum

# Tasks
# Merge online_retail_sales.csv with customer_details.csv using Customer_ID.
# Find the average order value by Membership_Level.
# Identify the membership group generating the highest revenue.
# Analyze purchasing behavior by age group.
```

```python
# Merge online_retail_sales.csv with customer_details.csv using Customer_ID.

merge_Task =pd.merge(data,df,on = "Customer_ID")

merge_Task.head(5)
```

```python
# Find the average order value by Membership_Level.

Avg_ord = merge_Task.groupby("Membership_Level")["Total_Amount"].mean().reset_index(name="AVG_Order_Value")

Avg_ord.head(6)
```

```python
# Identify the membership group generating the highest revenue.

high_rev = merge_Task.groupby("Membership_Level")["Revenue_Per_Item"].sum().reset_index(name = "Higest_Revenue")
high_rev=high_rev.sort_values(by ="Higest_Revenue", ascending=False)
high_rev.head(1)
```

```python
# Analyze purchasing behavior by age group.

age_behavior = merge_Task.groupby("Age")["Total_Amount"].sum().reset_index(name="Total_Spending")

age_behavior.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Visualize monthly revenue using a line chart.

Sales_by_Month = data.groupby('Month')['Revenue_Per_Item'].sum().sort_index()
Sales_by_Month

plt.figure(figsize = (18,5))
plt.plot(Sales_by_Month.index,Sales_by_Month.values,marker="o")
plt.title("Total Global Sales By Month")
plt.xlabel("Month")
plt.ylabel("Global Sales")
plt.grid(True)
plt.tight_layout()
# To Save --> plt.savefig("My_Chart.png")
plt.show()
```

```python
# Create a bar chart for revenue by category.

revenue_by_category = data.groupby("Category")["Revenue_Per_Item"].sum()

plt.figure(figsize=(10,6))
plt.bar(revenue_by_category.index, revenue_by_category.values)

plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.show()
```

```python
# Plot the top 10 customers by total spending.

top_10 = data.groupby("Customer_ID")["Total_Amount"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top_10.index.astype(str), top_10.values)

plt.title("Top 10 Customers by Total Spending")
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.tight_layout()
plt.show()
```
